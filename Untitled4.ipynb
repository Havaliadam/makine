```python
import zipfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
df_zip=zipfile.ZipFile("/content/archive.zip")

train=pd.read_csv(df_zip.open("train.csv"),low_memory=(False))
store=pd.read_csv(df_zip.open("store.csv"),low_memory=(False))

veri=train.merge(store,on=["Store"],how="inner")


veri["Date"]=pd.to_datetime(veri["Date"],infer_datetime_format=True)
veri["GÃ¼n"]=veri["Date"].dt.day
veri["Hafta"]=veri["Date"].dt.week
veri["Ay"]=veri["Date"].dt.month
veri["YÄ±l"]=veri["Date"].dt.year

veri["Mevsim"]=np.where(veri["Ay"].isin([3,4,5]),"ilkbahar",
np.where(veri["Ay"].isin([6,7,8]),"Yaz",
np.where(veri["Ay"].isin([9,10,11]),"sonbahar",
np.where(veri["Ay"].isin([12,1,2]),"KÄ±Å","None"


))))

veri[["Date","GÃ¼n","Hafta","Ay","YÄ±l","Mevsim"]]

veri["CompetitionDistance"]=veri["CompetitionDistance"].fillna(veri["CompetitionDistance"].mode()[0])
veri.isnull().sum()/veri.shape[0]*100


y=veri["Sales"]
nc=["Customers","Open","Promo","SchoolHoliday","CompetitionDistance","Promo2"]
kc=["DayOfWeek","StateHoliday","StoreType","Assortment","Hafta","Ay","YÄ±l","Mevsim"]
def encode_categorical(df, col):
    le = LabelEncoder()
    le1 = le.fit_transform(df[col]).reshape(-1, 1)
    oh = OneHotEncoder(sparse=False)
    column_names = [col + "_" + str(i) for i in le.classes_]
    return pd.DataFrame(oh.fit_transform(le1), columns=column_names)

t = veri[nc]
for column in kc:
    t_df = encode_categorical(veri, column)
    t = pd.concat([t, t_df], axis=1)

# Split data into train, test, and validation sets
x_train, x_test, y_train, y_test = train_test_split(t, y, test_size=0.2, random_state=0)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.1, random_state=0)

# Define the neural network model
model = Sequential()
model.add(Dense(64, input_dim=x_train.shape[1], activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(1, activation="linear"))  # linear activation for regression

# Compile the model
model.compile(optimizer="adam", loss="mae", metrics=["mean_absolute_error"])

# Train the model
model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=10)









#print("toplam maÄaza sayisi :{}".format(len(veri["Store"].unique())))
#print("gÃ¼nlÃ¼k ortalama satis :{}".format(round(veri["Sales"].mean(),2)))
# plt.figure(figsize=(15,8))
# sns.barplot(x="Mevsim",y="Sales",data=veri)
# # plt.hist(veri["Sales"])
# # plt.title("magaza satÄ±Å hist")
# # plt.xlabel("frenkans")
# plt.show()
```
